1. Basic Array Operations
Convert the mpg column into a NumPy array and calculate:
○ The mean, median, and standard deviation of mpg.
○ The number of cars with mpg greater than 25.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

array = df["mpg"].to_numpy()

mean_mpg = np.mean(array)
median_mpg = np.median(array)
std_mpg = np.std(array)

cars_gt_25 = np.sum(array > 25)

print("Mean MPG:", mean_mpg)
print("Median MPG:", median_mpg)
print("Standard Deviation MPG:", std_mpg)
print("Number of cars with MPG > 25:", cars_gt_25)
```

2. Filtering
Using NumPy, filter all cars with more than 6 cylinders.
Return the corresponding car_name as a list

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

cyl_array = df["cylinders"].to_numpy()
car_name_array = df["car name"].to_numpy()

filtered_cars = car_name_array[cyl_array > 6]

filtered_cars_list = filtered_cars.tolist()

print(filtered_cars_list)
```

3. Broadcasting
Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

hp_array = pd.to_numeric(df["horsepower"], errors="coerce").to_numpy()

mean_hp = np.nanmean(hp_array)
hp_array = np.where(np.isnan(hp_array), mean_hp, hp_array)

updated_hp = hp_array * 1.10

print(updated_hp)
```

4. Boolean Indexing
Find the average displacement of cars with an origin of 2 (Europe) using NumPy
indexing.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

origin_array = df["origin"].to_numpy()
displacement_array = df["displacement"].to_numpy()

europe_displacement = displacement_array[origin_array == 2]

avg_displacement = np.mean(europe_displacement)

print("Average Displacement (Europe cars):", avg_displacement)
```

5. Conditional Aggregates
Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("datafile.csv")

cyl_array = df["cylinders"].to_numpy()
mpg_array = df["mpg"].to_numpy()

unique_cyl = np.unique(cyl_array)

for cyl in unique_cyl:
    mean_val = np.mean(mpg_array[cyl_array == cyl])
    print(f"Cylinder {int(cyl)} : Mean MPG = {float(mean_val):.2f}")
```

2. Filtering and Indexing
Find all cars manufactured in 1975 with a weight less than 3000. Return the
DataFrame with selected columns: car_name, weight, and mpg.

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

filtered_df = df[(df["model year"] == 75) & (df["weight"] < 3000)]

result = filtered_df[["car name", "weight", "mpg"]]

print(result)
```

3. Handling Missing Data
Identify if there are any missing values in the dataset. Replace missing values in the
horsepower column with the column's median.

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

print("Missing values in each column:\n")
print(df.isnull().sum())

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

median_hp = df["horsepower"].median()
df["horsepower"].fillna(median_hp, inplace=True)

print("\nMissing values after replacement:\n")
print(df["horsepower"].isnull().sum())
```

4. Data Transformation
Add a new column power_to_weight_ratio, calculated as horsepower / weight.

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

df["power_to_weight_ratio"] = df["horsepower"] / df["weight"]

print(df[["car name", "horsepower", "weight", "power_to_weight_ratio"]].head())
```

5. Group By
Group the cars by origin and calculate the mean mpg for each group.

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

mean_mpg_origin = df.groupby("origin")["mpg"].mean()

print(mean_mpg_origin)
```

6. Visualization Preparation
Generate a summary DataFrame with:
○ Average mpg, weight, and horsepower for each model_year.

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

summary_df = df.groupby("model year")[["mpg", "weight", "horsepower"]].mean()

print(summary_df)
```

7. Exporting Data
Save a subset of the data containing only mpg, cylinders, horsepower, and weight
for cars with mpg > 30 into a CSV file named high_mpg_cars.csv.

```python
import pandas as pd

df = pd.read_csv("datafile.csv")

df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")

filtered_df = df[df["mpg"] > 30]

subset = filtered_df[["mpg", "cylinders", "horsepower", "weight"]]

subset.to_csv("high_mpg_cars.csv", index=False)

print("File saved successfully!")
```
